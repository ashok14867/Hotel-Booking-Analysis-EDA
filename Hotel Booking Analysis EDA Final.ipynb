````python
# we will deal with outliers using boxplot.

# In variable numeric_columns we stored the list of columns with numeric values

numeric_columns = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'required_car_parking_spaces',
           'adr', 'previous_cancellations', 'previous_bookings_not_canceled', 'booking_changes']

# We apply the for loop on numeric_columns to get outliers using boxplot

plt.figure(figsize=(18,12))
n = 1
for column in numeric_columns:
  plt.subplot(3,4,n)
  n = n+1
  sns.boxplot(df[column], color= 'Red')
plt.show()
````

#### Chart - 2

#### Preference of hotel type by visitors in top 10 countries visited

````python
# Separate city hotels
type_city = df['hotel'] == 'City Hotel'

# Sort top 10 countries according to city hotels
sort_city = df[type_city]['country'].value_counts(normalize=True).head(10).reset_index()
sort_city.columns = ['country', 'proportion']

# Sort top 10 countries according to resort hotels
sort_resort = df[~type_city]['country'].value_counts(normalize=True).head(10).reset_index()
sort_resort.columns = ['country', 'proportion']

# Add a column for hotel type
sort_city['hotel'] = 'City Hotel'
sort_resort['hotel'] = 'Resort Hotel'

# Concatenate the dataframes
df3 = pd.concat([sort_city, sort_resort]).reset_index(drop=True)

# Check the structure of df3
print(df3.head())

# Plot the bar plot with count and hotel types
plt.figure(figsize=(14, 8))
sns.barplot(x='country', y='proportion', data=df3, hue='hotel', palette='muted')
plt.title('Top 10 Countries by Hotel Type')
plt.xlabel('Country')
plt.ylabel('Proportion of Bookings')
plt.xticks(rotation=45)
plt.legend(title='Hotel Type')
plt.show()
````

Choosing this comparison allows us to draw meaningful conclusions about travel behaviors and preferences, which can be invaluable for market analysis, business development, and targeted marketing efforts.
